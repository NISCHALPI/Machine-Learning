```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib.image as mpimg # .jpg .png 
```

```python
image_array = mpimg.imread(fname='../DATA/palm_trees.jpg')
```

```python
h,w,c = image_array.shape
```

```python
image_array = image_array.reshape(h*w,c)
```

```python
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=6, n_init='auto', random_state=101)
lables = cluster.fit_predict(image_array)
```

```python
centers = cluster.cluster_centers_.round(0)
```

```python
#Quantized Image
plt.imshow(centers[lables].reshape(h,w,c).astype(int))
```
