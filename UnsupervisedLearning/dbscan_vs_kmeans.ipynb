```python
circles = pd.read_csv(filepath_or_buffer='./notebooks/DATA/cluster_circles.csv')
```

```python
sns.scatterplot(data=circles , x='X1', y='X2')
```

```python
def display_categories(model , data_):
    lbl = model.fit_predict(data_)
    sns.scatterplot(data=data_ , x='X1', y='X2', hue=lbl, palette='Set1')
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
display_categories(KMeans(n_clusters=3, n_init='auto', max_iter=10**6), blobs)
```

```python
display_categories(KMeans(n_clusters=2, n_init='auto', max_iter=10**6), moons)
```

```python
display_categories(KMeans(n_clusters=2, n_init='auto', max_iter=10**6), circles)
```

```python
display_categories(DBSCAN(n_jobs=-1), blobs)
```

```python
display_categories(DBSCAN(eps=0.2), moons)
```

```python
display_categories(DBSCAN(eps=0.2, n_jobs=-1), circles)
```
